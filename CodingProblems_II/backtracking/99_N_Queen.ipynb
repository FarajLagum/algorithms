```python
# Define the size of the board
N = 1

# Initialize the board with zeros
board = [[0 for i in range(N)] for j in range(N)]

# A function to check if a position is safe for a queen
def is_safe(row, col):
    # Check the same row
    for i in range(col):
        if board[row][i] == 1:
            return False
    # Check the upper left diagonal
    i = row
    j = col
    while i >= 0 and j >= 0:
        if board[i][j] == 1:
            return False
        i -= 1
        j -= 1
    # Check the lower left diagonal
    i = row
    j = col
    while i < N and j >= 0:
        if board[i][j] == 1:
            return False
        i += 1
        j -= 1
    # If none of the above conditions are violated, the position is safe
    return True

# A recursive function to solve the N Queens Problem
def solve_n_queen(col):
    # If all queens are placed, return True
    if col == N:
        return True
    # Try placing a queen in each row of the current column
    for i in range(N):
        # Check if the position is safe
        if is_safe(i, col):
            # Place the queen
            board[i][col] = 1
            # Recur for the next column
            if solve_n_queen(col + 1):
                return True
            # If placing the queen does not lead to a solution, backtrack
            board[i][col] = 0
    # If no solution is found, return False
    return False

# A function to print the board
def print_board():
    for i in range(N):
        for j in range(N):
            print(board[i][j], end=" ")
        print()

# Call the solve function and print the board if a solution exists
if solve_n_queen(0):
    print_board()
else:
    print("No solution exists")
```

```python
def is_safe(board, row, col, n):
    # Check if there is a queen in the same column
    for i in range(row):
        if board[i][col] == 1:
            return False

    # Check if there is a queen in the upper-left diagonal
    for i, j in zip(range(row, -1, -1), range(col, -1, -1)):
        if board[i][j] == 1:
            return False

    # Check if there is a queen in the upper-right diagonal
    for i, j in zip(range(row, -1, -1), range(col, n)):
        if board[i][j] == 1:
            return False

    return True

def print_solution(board):
    for row in board:
        print(" ".join(map(str, row)))
    print()

def solve_n_queens_util(board, row, n):
    if row == n:
        # All queens are placed successfully
        print_solution(board)
        return

    for col in range(n):
        if is_safe(board, row, col, n):
            # Place queen and move to the next row
            board[row][col] = 1

            # Recur for the next row
            solve_n_queens_util(board, row + 1, n)

            # Backtrack: Remove the queen from the current cell
            board[row][col] = 0

def solve_n_queens(n):
    board = [[0] * n for _ in range(n)]
    solve_n_queens_util(board, 0, n)

# Example usage:
n = 8  # You can change the value of 'n' to solve for different board sizes
solve_n_queens(n)
```

```python
def is_safe(board, row, col, n):
    for i in range(row):
        if board[i][col] == 1:
            return False

    for i, j in zip(range(row, -1, -1), range(col, -1, -1)):
        if board[i][j] == 1:
            return False

    for i, j in zip(range(row, -1, -1), range(col, n)):
        if board[i][j] == 1:
            return False

    return True

def write_solution_to_file(file, board):
    for row in board:
        file.write(" ".join(map(str, row)) + "\n")
    file.write("\n")

def solve_n_queens_util(board, row, n, solution_file):
    if row == n:
        # All queens are placed successfully, write the solution to the file
        write_solution_to_file(solution_file, board)
        return

    for col in range(n):
        if is_safe(board, row, col, n):
            board[row][col] = 1
            solve_n_queens_util(board, row + 1, n, solution_file)
            board[row][col] = 0

def solve_n_queens_and_save_to_file(n):
    board = [[0] * n for _ in range(n)]
    with open("n_queens_solutions.txt", "w") as solution_file:
        solve_n_queens_util(board, 0, n, solution_file)

# Example usage:
n = 8  # You can change the value of 'n' to solve for different board sizes
solve_n_queens_and_save_to_file(n)
```
